# Extracting the words around one specific word in the corpus

```python
def sentences_in_file(fname,word):
    f=open(PATH_PLAIN + fname,'r')
    sentences_list = f.read().split('.')
    sentences_with_word = []
    for sentence in sentences_list:
        if word in sentence:
            sentences_with_word.append(sentence)
    return sentences_with_word

def words_in_same_sentence_in_corpus(word):
    sentences = ''
    words_count={}
    for fname in LIST_PLAIN:
        sentences_with_word=sentences_in_file(fname,word)
        if sentences_with_word:
            for sentence in sentences_with_word:
                if word in [word.strip(string.punctuation).lower() for word in sentence.split()]:
                    sentences += (' '+sentence)
    for word in [word.strip(string.punctuation).lower() for word in sentences.split()]:
        words_count[word] = words_count.get(word,0) + 1
    return words_count
```

```python
around_excellent = words_in_same_sentence_in_corpus('excellent')
```
